```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
fastfood = pd.read_csv('Downloads/fastfood.csv')
```

```python
fastfood.head()
```

```python
fastfood.isnull().sum()
```

```python
fastfood.corr()
```

```python
fastfood.describe()
```

```python
#   
sns.heatmap(fastfood.corr(),annot=True)
```

```python
nutritions_per_restaurant = fastfood.groupby('restaurant').mean().round(2).sort_values(by='calories').reset_index()
nutritions_per_restaurant
```

```python
#      
fig, ax = plt.subplots(figsize=(15, 8))

custom_palette = ['#D0D0D0'] *(len(nutritions_per_restaurant['restaurant'].unique())-2) + ['#FF7533']
sns.barplot(x = 'trans_fat', y='restaurant', data=nutritions_per_restaurant, ax=ax
            , palette=custom_palette)
plt.title('Highest trans fat Median Based on Restaurant', fontsize=15)
plt.xlabel('trans_fat', fontsize=10)
plt.ylabel('')

for i in ax.containers:
    plt.bar_label(i, label_type='edge')

plt.show()
```
